```python
movie_pie = df["type"].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.pie(movie_pie, labels=movie_pie.index, autopct='%1.1f%%' , startangle=190, colors=['#ff9999','#66b3ff'])
plt.title("Type")
plt.axis('equal')  # Grafiği daire şeklinde yapar
plt.show()
```

```python
df['director'].fillna("Unknown", inplace = True)
df["cast"].fillna("Unknown", inplace = True)
df["country"].fillna("Unknown", inplace = True)
df["date_added"].fillna("Unknown", inplace = True)
df["rating"].fillna("Unrating", inplace = True)
df
```

```python
plt.figure(figsize=(10,6))
top_genres = df['listed_in'].str.split(', ').explode().value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index, palette="viridis")
plt.title('Top 10 Genres on Netflix', fontsize=14)
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
plt.figure(figsize = (6,6))
country = df["country"].str.split(', ').explode().value_counts().head(10)
sns.barplot(x=country.values, y=country.index, palette="mako")
plt.title("Top 10 Country on Netflix Movies", fontsize = 14)
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```
